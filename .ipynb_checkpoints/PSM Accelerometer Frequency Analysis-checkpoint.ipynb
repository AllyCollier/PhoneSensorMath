### GPS Tracking Analysis ###

This jupyter notebook is designed to accept a comma-separated (csv) list of timestamped GPS latitude and longitude coordinates representing a recording of a walking or running exercise and find the fastest mile-long segment within the longer exercise. While it is taylored to work with the output from the __[Physics Toolbox Sensor Suite app](https://www.vieyrasoftware.net/)__, we have tried to make the instructions here agnostic to the method of GPS recording used.

#### The Input File Specification ####
What follows assumes that the input .csv file resembles:
```
time,Latitude,Longitude,,
08:29:14:458,37.38775891,-6.00309754,
08:29:15:460,37.38774677,-6.00311591,
08:29:16:448,37.38774284,-6.00313017,
08:29:18:468,37.38775696,-6.00314697,
08:29:19:463,37.38776887,-6.00314499,

````python
import pandas as pd
import time
import numpy as np
from numpy.fft import fft
import matplotlib
import matplotlib.pyplot as plt

# Use LaTeX computer modern font
from matplotlib import rc
rc('font', **{'family': 'serif', 'serif': ['Computer Modern']})
rc('text', usetex=True)
````

#### Analysis of washer vibrations during spin cycle. Sevilla.

Accelerometer readings during spin cycle of washer in Sevilla flat. Washer advertises a 1200RPM spin cycle. Measured using Physics Toolbox Sensor Suite Pro, OnePlus 3T, lying atop the washer, x-axis perpendicular to front of washer. Device seems capable of sampling forces at about 200Hz. Should be able to detect 1200RPM (60Hz) vibrations.

````python
# Read the accelerometer data
raw = pd.read_csv('Washer-2018-10-1313.10.26.csv')
````

````python
# Compute fft on interval [p, p+l]
# Empirically, it seems that gFx is the axis parallel to the width of the phone.
# By inspecting raw data: time is 0 to 1856 seconds
# Compute peak frequency on 10 sec intervals in steps of 10 seconds.

# Total number of samples in recording
num_samples = raw.shape[0]

# Length of window, seconds
l = 5

# Grab the sample times
t = np.array(raw.time)

# Grab the sample values
g = np.array(raw.ay)

# Full fft for this window
frequencies = []

# Modes to keep, e.g. since this device samples maximally at 200Hz, no point keeping more than 200 freqs for each window.
num_modes_to_keep = 100

for p in range(0,num_samples):
    # break if too close to the end of the sample
    if t[-1]-p < l:
        break;
    # create l sec window mask
    w = (p<t) & (t<l+p)
    # do fft on the data in that window
    f = np.abs(fft(g[w]))
    # add f to list of frequencies
    frequencies.append(f[1:num_modes_to_keep])

#plt.bar(range(1,100),frequencies[100][1:100])
#plt.show()
````

````python
fig, ax = plt.subplots(figsize=[5.5, 4.125])
ax.set_ylabel('Spin Frequency (Hz)')
ax.set_yticks([i for i in range(0,num_modes_to_keep+10,10)])
ax.set_yticklabels(["{:.1f}".format(i/l) for i in range(0,num_modes_to_keep+10,10)])
ax.set_xlabel('Time (min)')
ax.set_ylim((0,110))
ax.set_xticks([i for i in range(0,510,60)])
ax.set_xticklabels(["{:.0f}".format(i/60) for i in range(0,510,60)])
im = ax.imshow(np.transpose(frequencies),aspect='auto',origin='lower',cmap='gray_r')
ax.grid()
plt.savefig("washer-ay.eps")
````
